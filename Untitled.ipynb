```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
products_df = pd.read_csv('products.csv')
users_df = pd.read_csv('users.csv')
orders_df = pd.read_csv('order_quantity.csv')
reviews_df = pd.read_csv('reviews.csv')

# Display the first few rows of each DataFrame to inspect the data
print("Products DataFrame:")
print(products_df.head())
print("\nUsers DataFrame:")
print(users_df.head())
print("\nOrders DataFrame:")
print(orders_df.head())
print("\nReviews DataFrame:")
print(reviews_df.head())
```

```python
# Merge products and orders to calculate sales data
sales_df = pd.merge(orders_df, products_df, on='product_id')

# Calculate the total price for each order line item
sales_df['total_price'] = sales_df['quantity'] * sales_df['price']

# Merge the sales data with user information
full_df = pd.merge(sales_df, users_df, on='user_id')

# Convert the 'order_date' column to datetime objects
full_df['order_date'] = pd.to_datetime(full_df['order_date'])

# Display the first few rows of the merged DataFrame
print("\nMerged DataFrame:")
print(full_df.head())
```

```python
# Group by product name and sum the total price
top_selling_products = full_df.groupby('product_name')['total_price'].sum().sort_values(ascending=False)

# Display the top 5 selling products
print("\nTop 5 Selling Products:")
print(top_selling_products.head())

# Create a bar chart for visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=top_selling_products.index, y=top_selling_products.values, palette='viridis')
plt.title('Top-Selling Products by Total Revenue', fontsize=16)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Total Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()# Group by product name and sum the total price
```

```python
# Extract the hour from the order_time column
full_df['order_hour'] = pd.to_datetime(full_df['order_time']).dt.hour

# Count the number of orders for each hour
peak_hours = full_df['order_hour'].value_counts().sort_index()

# Create a line chart to visualize peak hours
plt.figure(figsize=(12, 6))
sns.lineplot(x=peak_hours.index, y=peak_hours.values, marker='o', color='b')
plt.title('Number of Orders by Hour of the Day', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(range(24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Count the number of orders per user
user_orders = full_df.groupby('username')['order_id'].count().sort_values(ascending=False)

# Display the top 5 most active users
print("\nTop 5 Most Active Users:")
print(user_orders.head())

# Create a bar chart to show user activity
plt.figure(figsize=(10, 6))
sns.barplot(x=user_orders.index, y=user_orders.values, palette='rocket')
plt.title('Number of Orders per User', fontsize=16)
plt.xlabel('Username', fontsize=12)
plt.ylabel('Total Number of Orders', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
products_df = pd.read_csv('products.csv')
reviews_df = pd.read_csv('reviews.csv')

# Merge products and reviews data
reviews_with_products = pd.merge(reviews_df, products_df, on='product_id')

# Calculate the average rating for each product, grouped by category for the hue
average_ratings_with_category = reviews_with_products.groupby(['category', 'product_name'])['rating'].mean().reset_index()

# Create a bar chart to visualize average ratings, using hue for category
plt.figure(figsize=(12, 7))
sns.barplot(x='product_name', y='rating', hue='category', data=average_ratings_with_category, palette='viridis')
plt.title('Average Product Ratings by Category', fontsize=16)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 5)
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```
